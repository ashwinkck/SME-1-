1) Importing the tools

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from  langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from groq import Groq
```

2)Loading the PDF

```python
loader = PyPDFLoader(r"C:\Users\ashwi\Desktop\medicaps\langchain\data\EJ1172284.pdf")
documents = loader.load()
```

3)Splitting the document

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 50
)
chunks = text_splitter.split_documents(documents)
```

4)Creating the embeddings

```python
embeddings = HuggingFaceEmbeddings(
    model_name = "sentence-transformers/all-MiniLM-L6-v2"

)
```

5) Building the Vector Database

```python
vector_store = FAISS.from_documents(
    chunks,
    embeddings
)
```

6) Initialise the LLM

```python
client = Groq(api_key="api_key")
```

7. Interactive Question Loop

```python
while True:
    question = input("Ask the qn:")

    if question.lower() == "exit":
        break

    docs = vector_store.similarity_search(
        question,
        k=3
    )
    context = "\n\n".join([doc.page_content for doc in docs])

    #Prompt Engineering
    prompt = f"""
    use the following context to answer the question.
Context:
{context}

Question:
{question}

Answer clearly
"""
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "user","content":prompt}
        ]
    )

    # Extracting the Answer
    answer = response.choices[0].message.content
    print("\n Ai answer: \n",answer)
```
